```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("rainfall_timeseries.csv")

#View first rows
df.head()
```

```python
# Combine Year and Month into a date
df['Date'] = pd.to_datetime(df[['Year', 'Month']].assign(DAY=1))

# Sort values
df = df.sort_values('Date')

# Plot rainfall over time
plt.figure()
plt.plot(df['Date'], df['Precipitation'], marker='o')
plt.xlabel("Time")
plt.ylabel("Rainfall (mm)")
plt.title("Rainfall Over Time")
plt.show()
```

```python
plt.figure()
plt.plot(df['Date'], df['Temperature'], label="Temperature")
plt.plot(df['Date'], df['Precipitation'], label="Rainfall")

plt.xlabel("Time")
plt.ylabel("Values")
plt.title("Temperature vs Rainfall Over Time")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.scatter(df["Precipitation"], df["Temperature"])

plt.xlabel("Rainfall")
plt.ylabel("Temperature")
plt.title("Rainfall vs Temperature")
plt.show()
```
